```python
import os, glob, cv2, matplotlib.pyplot as plt
import random

# Correct recursive paths
english_images = glob.glob(r"D:\datasets_ml\iam_words\iam_words\words\**\*.png", recursive=True)
telugu_images  = glob.glob(r"E:\IIIT-HW-Telugu_v1\TeluguSeg\**\*.jpg", recursive=True)

print("English samples:", len(english_images))
print("Telugu samples:", len(telugu_images))

# Pick random 3 from each for visualization
if english_images and telugu_images:
    plt.figure(figsize=(10,6))
    
    # English samples
    for i, img_path in enumerate(random.sample(english_images, 3)):
        plt.subplot(2,3,i+1)
        plt.imshow(cv2.imread(img_path), cmap='gray')
        plt.title(f"English {i+1}")
        plt.axis('off')

    # Telugu samples
    for i, img_path in enumerate(random.sample(telugu_images, 3)):
        plt.subplot(2,3,3+i+1)
        plt.imshow(cv2.imread(img_path), cmap='gray')
        plt.title(f"Telugu {i+1}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

else:
    print("No images found — please double-check dataset paths or extraction.")
```
